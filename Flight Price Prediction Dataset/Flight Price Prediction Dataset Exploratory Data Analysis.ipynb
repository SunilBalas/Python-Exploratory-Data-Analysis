## Feature Engineering

```python
# Date of Journey Feature

df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]

df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Arrival Time Feature

df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0])

df['Arrival_Hours'] = df['Arrival_Time'].apply(lambda x: x.split(':')[0])
df['Arrival_Minutes'] = df['Arrival_Time'].apply(lambda x: x.split(':')[1])

df['Arrival_Hours'] = df['Arrival_Hours'].astype(int)
df['Arrival_Minutes'] = df['Arrival_Minutes'].astype(int)

df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Departure Time Feature

df['Dep_Time'] = df['Dep_Time'].apply(lambda x: x.split(' ')[0])

df['Dep_Hours'] = df['Dep_Time'].apply(lambda x: x.split(':')[0])
df['Dep_Minutes'] = df['Dep_Time'].apply(lambda x: x.split(':')[1])

df['Dep_Hours'] = df['Dep_Hours'].astype(int)
df['Dep_Minutes'] = df['Dep_Minutes'].astype(int)

df.drop('Dep_Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
# Total Stops Feature

df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4})
```

```python
df.head()
```

```python
df.drop('Route', axis=1, inplace=True)
```

```python
df.drop([6474, 2660], axis=0, inplace=True)
```

```python
# Duration Feature

def hours_to_minutes(duration):
    duration_arr = duration.split(' ')
    hour = duration_arr[0].replace('h', '')
    
    if len(duration_arr) > 1:
        minute = duration_arr[1].replace('m', '')
        return int(hour) * 60 + int(minute)
    return int(hour) * 60
```

```python
df['Duration'] = df['Duration'].apply(hours_to_minutes)
```

```python
df.head()
```

```python
df['Airline'].unique()
```

```python
# Label Encoding

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
```

```python
df['Airline'] = encoder.fit_transform(df['Airline'])
df['Source'] = encoder.fit_transform(df['Source'])
df['Destination'] = encoder.fit_transform(df['Destination'])
df['Additional_Info'] = encoder.fit_transform(df['Additional_Info'])
```

```python
# One Hot Encoding

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
```

```python
df['Airline'] = ohe.fit_transform(np.array(df['Airline']).reshape(-1,1)).toarray()
df['Source'] = ohe.fit_transform(np.array(df['Source']).reshape(-1,1)).toarray()
df['Destination'] = ohe.fit_transform(np.array(df['Destination']).reshape(-1,1)).toarray()
df['Additional_Info'] = ohe.fit_transform(np.array(df['Additional_Info']).reshape(-1,1)).toarray()
```
