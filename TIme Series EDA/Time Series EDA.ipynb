## Time Resampling

```python
df_tesla.head()
```

```python
# rule = 'A' --> Year End Frequency
df_tesla.resample(rule='A').min()
```

```python
df_tesla.resample(rule='A').max()
```

```python
type(df_tesla.resample(rule='A').min())
```

```python
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
# Quarterly Start
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
# Business End Frequency

df_tesla.resample(rule='BA').max().head()
```

```python
# Business Quarter Start Frequency

df_tesla.resample(rule='BQS').max().head()
```

```python
df_tesla['Open'].resample(rule='BA').mean().plot(kind='bar')
```

```python
df_tesla['High'].resample(rule='YS').min().plot()
```

```python
import seaborn as sns

sns.histplot(df_tesla['Open'].resample(rule='QS').max(), kde=True)
```
